## seaborn:分布绘图-单变量绘图

内容介绍：分为单变量，双变量及多变量的分布绘图。

* 主要使用distplot()函数，结合kind参数进行绘图。
* 最新版本中，distplot在未来将被移除，被displot代替。

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#加载数据
```

```python
# 网络加载官方案例数据
# tips = sns.load_dataset('tips')

#使用本地数据加载
df = pd.read_csv('seaborn-data/titanic.csv')
df.head()
```

### 1.单变量分布

主要采取直方图的形式绘制

```python
# 通过图形查看年龄分布
# 去除数据中的nan
df1 = df[~np.isnan(df['age'])]
df1.head()
```

### 1.1使用distplot函数

```python
#首先设置颜色盘
sns.set(color_codes=True)
#默认绘制的y轴为比例，而不是实际数量
sns.distplot(df1['age'])
```

```python
#首先设置颜色盘
sns.set(color_codes=True)
#设置kde=False，显示实际数量
sns.distplot(df1['age'], kde=False)
```

### 1.2使用displot函数

```python
#首先设置颜色盘
sns.set(color_codes=True)
#设置kde=False，显示实际数量。ked为“核密度”曲线参数。
sns.displot(df1['age'], kde=True)
```

```python
#首先设置颜色盘
sns.set(color_codes=True)
#设置kde=False，显示实际数量
sns.displot(df1['age'])
```

```python
#首先设置颜色盘
sns.set(color_codes=True)
# bins参数，设置直方图的数量。默认bins=20,即20条直方图。
sns.displot(df1['age'],kde=True,bins=30)
```

```python
#首先设置颜色盘
sns.set(color_codes=True)
# rug参数，显示数据的集中程度。
sns.displot(df1['age'],kde=True,bins=30,rug=True)
```

```python
#首先设置颜色盘
sns.set(color_codes=True)
# hist=False参数，关掉直方图。注意displot()函数不能使用hist参数关掉直方图。
sns.distplot(df1['age'],kde=True,bins=30,rug=True,hist=False)
```

### 1.3单独使用kde函数绘制曲线

```python
#使用shade=True参数，是曲线下覆盖的面积显示阴影。
#根据kde函数计算原则，曲线下覆盖的面积=1.
sns.kdeplot(df['age'],shade=True)
```

KDE 的带宽参数（bw）控制着密度估计曲线的宽窄形状，有点类似直方图中的 bins 参数的作用。它对应着我们上面绘制的 KDE 的宽度。默认情况下，函数会按照一个通用的参考规则来估算出一个合适的值，但是尝试更大或者更小也可能会有帮助：

```python
#多个带宽同时绘制
x=df['age']
sns.kdeplot(x)
sns.kdeplot(x, bw_method=.2, label="bw: 0.2")
sns.kdeplot(x, bw_method=2, label="bw: 2")
plt.legend();
```
