## seaborn:FaceGrid技术讲解

内容介绍：

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#加载数据
```

```python
# 网络加载官方案例数据
# tips = sns.load_dataset('tips')

#使用本地数据加载
df = pd.read_csv('seaborn-data/tips.csv')
df.head()
```

### 1.FacetGrid图像绘制的基本方法

```python
#形成FaceGrid对象。比较简单，美化不足。
g = sns.FacetGrid(df)
#传入绘图函数plt.scatter。传入的参数要求：（1）传入的函数没有使用FaceGrid方法,（2）传入函数可以使用color参数
g.map(plt.scatter, 'total_bill', 'tip')
```

### 2.绘制多个图形

```python
#传入col参数，col的数据中包含的单一值个数应为合适的数个，不能较多。
g2 = sns.FacetGrid(df,col='day')
g2.map(plt.scatter, 'total_bill', 'tip')
```

```python
#也可以传入row参数
g2 = sns.FacetGrid(df,col='day',row='time')
g2.map(plt.scatter, 'total_bill', 'tip')
```

### 3.主要参数

```python
#添加hue参数
g2 = sns.FacetGrid(df,col='day',hue='time')
g2.map(plt.scatter, 'total_bill', 'tip')
#添加图例
g2.add_legend()
```

```python
g = sns.FacetGrid(df,col='smoker')
g.map(sns.barplot, 'day','total_bill',order=['Thur','Fri','Sta','Sun'])
```

```python
#设置图像大小
g = sns.FacetGrid(df,col='smoker',height=4,aspect=2)
g.map(sns.barplot, 'day','total_bill',order=['Thur','Fri','Sta','Sun'])
```
