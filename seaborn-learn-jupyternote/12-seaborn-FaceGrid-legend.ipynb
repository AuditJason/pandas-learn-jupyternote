## seaborn:FaceGrid设置图例等绘图细节

内容介绍：

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#加载数据
```

```python
# 网络加载官方案例数据
# tips = sns.load_dataset('tips')

#使用本地数据加载
df = pd.read_csv('seaborn-data/tips.csv')
df.head()
```

### 1.legend指定参数

```python
#支持中文。rc参数给matplotlib设置参数。
sns.set(rc={'font.sans-serif':'AR PL UKai CN'})
#添加hue参数
g2 = sns.FacetGrid(df,col='day',hue='time')
g2.map(plt.scatter, 'total_bill', 'tip')
#添加图例。返回警告，不推荐使用legend传入的参数形式。
g2.add_legend(title='标题',labels=['午餐','晚餐'],label_order=['Lunch','Dinner'])
```

### 2.设置子图例的标题

set_titles(template=None, row_template=None, col_template=None, **kwargs) method of seaborn.axisgrid.FacetGrid instance
    Draw titles either above each facet or on the grid margins.
    
    Parameters
    ----------
    template :  {col_var} and  {col_name} and/or {row_var} and {row_name} 
    row_template:Must have {row_var} and {row_name} formatting keys.
    col_template:Must have {col_var} and {col_name} formatting keys.
    参数{}外可以添加文字说明。

```python
g2 = sns.FacetGrid(df, col='day', row='time')
g2.map(plt.scatter, 'total_bill', 'tip')
#为每个子图添加图例。template='{变量名/变量值}+任意必要字符串'
#g2.set_titles(template='{row_var}/{row_name}')
g2.set_titles(template='{col_var}/{col_name}')
#g2.set_titles(template='时间{row_name}/星期{col_name}')
```

### 3.设置坐标轴信息

```python
#set_axis_labels方法批量设置横纵坐标轴标题
g2 = sns.FacetGrid(df, col='day', row='time')
g2.map(plt.scatter, 'total_bill', 'tip')
g2.set_axis_labels('消费金额','小费')
```

```python
#g2.set方法更多的图形信息
g2 = sns.FacetGrid(df, col='day', row='time')
g2.map(plt.scatter, 'total_bill', 'tip')
# set方法能够设置的属性，同matplotlib中的axis类，可详细查看matplotlib官方文档学习。
#g2.set(facecolor='y')
g2.set(xticks=range(0,50,10),xticklabels=['${}'.format(x) for x in range(0,50,10)])
```

### 4.设置图像信息

```python
#g4.fig方法更多的图形信息
g4 = sns.FacetGrid(df, col='day', row='time')
g4.map(plt.scatter, 'total_bill', 'tip')
#通过fig方法中色set_dpi设置图像分辨率
# fig方法能够设置的属性，同matplotlib中的axis类中的相关方法，可详细查看matplotlib官方文档学习。
g4.fig.set_dpi(100)
```
