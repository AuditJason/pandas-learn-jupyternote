## seaborn:成对绘图

内容介绍：可以把某个数据集合中几个字段之间的关系一次性绘制出来。

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
#加载数据
```

```python
# 网络加载官方案例数据
# tips = sns.load_dataset('tips')

#使用本地数据加载-鸢尾花数据集
df = pd.read_csv('seaborn-data/iris.csv')
df.head()
```

```python
#使用本地数据加载-鸢尾花数据集
df2 = pd.read_csv('seaborn-data/tips.csv')
df2.head()
```

### 1.绘图基本方式，及设置比较数据列

```python
#所有数值类型的列，进行绘图
# 对角线是直方图，x轴=y轴，表示当前列的分布情况。
sns.pairplot(df)
```

```python
# vars参数设置需要绘制出来的字段。
sns.pairplot(df2,vars=['total_bill','tip'])
```

### 2.设置图形列别

```python
#kind参数，主图形的形式
#diag_kind参数，斜对角图形的形式
sns.pairplot(df, vars=['sepal_length', 'sepal_width', 'petal_length'],  kind='reg', diag_kind='kde')
```
