## seaborn:二变量分布图

内容介绍：

* 可以使用散点图
* 也可使用散点图结合直方图绘制
* 使用jointplot()函数进行。比传统的三点图增加在底部上上部增加直方图。

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
#加载数据
```

```python
# 网络加载官方案例数据
# tips = sns.load_dataset('tips')

#使用本地数据加载
df = pd.read_csv('seaborn-data/tips.csv')
df.head()
```

### 1.jointplot()基本使用

```python
sns.jointplot(x='total_bill', y='tip', data=df)
```

### 2.jointplot()主要参数

```python
#kind='reg',出现kde曲线和回归线
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
```

```python
#加载各国运动员的数据
df2 = pd.read_csv('otherdata/athlete_events.csv')
df3 =df2[df2['NOC']=='CHN']
df3.head()
```

```python
#数据量较大时使用六边形图，使数据区域变成六边形，颜色越深数据分布越集中。
# gridsize=15，六边形的数量。六边形越少，显示越大。
#首先设置颜色盘
sns.set(color_codes=True)
sns.jointplot(x='Height', y='Weight', data=df2, kind='hex', gridsize=25)
```

### 3.其他参数

* color: 绘制元素的颜色
* height: 图的大小，图是一个正方形
* ratio: 主图和副图的比例，只能为一个整型
* space: 主图和幅图的间距

```python
#color设置颜色样式
sns.set(color_codes=True)
sns.jointplot(x='Height', y='Weight', data=df2, kind='hex', gridsize=25, color='red')
```

```python
#height=10设置图形大小
sns.set(color_codes=True)
sns.jointplot(x='Height', y='Weight', data=df2, kind='hex', gridsize=25, height=10)
```

```python
#ratio设置主图幅图的比例。默认为ratio=5,表示主图：副图=5：1
sns.set(color_codes=True)
sns.jointplot(x='Height', y='Weight', data=df2, kind='hex', height=10,gridsize=25,  ratio=4)
```

```python
#space=0调整主幅图之间的间距为0。
sns.set(color_codes=True)
sns.jointplot(x='Height', y='Weight', data=df2, kind='hex', height=10,gridsize=25,  ratio=4,space=0)
```

```python
#marginal_kws设置幅图上的属性。
sns.set(color_codes=True)
sns.jointplot(x='Height', y='Weight', data=df2, kind='hex', height=10, gridsize=25,  ratio=4, space=0, marginal_kws={'rug':True, 'kde':True})
```
