## seaborn:分类图,分布散点图和蜂群图

内容介绍：使用sns.catplot()，不同类别的图，使用kind参数指定


* (1)strplot():使用sns.catplot(kind='strip'),也是catplot的默认值。
* (2)swarmplot():sns.catplot(kind='swarm')。

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

### 加载数据：

```python
# 网络加载官方案例数据
# tips = sns.load_dataset('tips')

#使用本地数据加载
df = pd.read_csv('seaborn-data/tips.csv')
df.head()
```

### 1.分布散点图

```python
#（1）strplot():
#简单示例
sns.catplot(x='day',y='total_bill', data=df)
```

```python
#catplot默认strplot()，可以不添加kind参数
sns.catplot(x='day',y='total_bill', data=df, hue='sex')
```

### 2.蜂群分布图

```python
#（2）swarmplot()，显示的样式不同，可以看出分布，让点不重叠。
# 此种方式适合小型数据类型。大型数据排不下
sns.swarmplot(x='day',y='total_bill', data=df)
```

```python
#使用catplot+kind参数的形式
sns.catplot(x='day',y='total_bill', data=df, kind='swarm')
```

```python
#添加hue参数
sns.catplot(x='day',y='total_bill', data=df, kind='swarm',hue='sex')
```
