## seaborn:自带样式和风格

内容介绍：自带5种风格，分别是

* white,白色
* whitegrid,带有网格的白色
* dark,灰色
* darkgrid,灰色带网格(网格白色)
* ticks,白色、并在轴上带有刻度条

**共有三种函数设置颜色**
* 使用set_style方法进行设置
* 使用axis_style设置
* 使用set设置

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
#加载数据
```

```python
# 网络加载官方案例数据
# tips = sns.load_dataset('tips')

#使用本地数据加载
df = pd.read_csv('seaborn-data/tips.csv')
df.head()
```

### 1.使用set_style设置

* 设置的是全局样式，其他部分绘制的图都采用此样式

```python
# white
sns.set_style('white')
axes = sns.scatterplot(x='total_bill',y='tip',data=df)
```

```python
# whitegrid
sns.set_style('whitegrid')
axes = sns.scatterplot(x='total_bill',y='tip',data=df)
```

```python
# drak
sns.set_style('dark')
axes = sns.scatterplot(x='total_bill',y='tip',data=df)
```

```python
# darkgrid
sns.set_style('darkgrid')
axes = sns.scatterplot(x='total_bill',y='tip',data=df)
```

```python
# ticks：xy轴上有刻度线
sns.set_style('ticks')
axes = sns.scatterplot(x='total_bill',y='tip',data=df)
```

### 2.sns.axes_styles

* 只能临时和局部设置。使用with代码快进行
* 设置的属性较多

```python
#使用本地数据加载
df2 = pd.read_csv('seaborn-data/fmri.csv')
df2.head()
```

```python
#通过axes_style中的rc参数，可以设置其他各种样式要素
with sns.axes_style('darkgrid',rc={'axes.facecolor':'y', 'grid.color':'black'}):
    axes = sns.lineplot(x='timepoint',y='signal',data=df2)
```

```python
#脱离代码快后，原来设置的样式不能保存
axes = sns.lineplot(x='timepoint',y='signal',data=df2)
```

### 3.set()方法。

* 比较全面的方法。具体使用需要参考matplotlib中的相关说明进行设置。
* 根据帮助文件，这个函数会在未来取消。

```python
#set函数中会针对样式存在默认值，不进行任何设置也能够出现样式。
sns.set()
axes = sns.lineplot(x='timepoint',y='signal',data=df2)
```
