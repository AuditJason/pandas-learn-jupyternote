## seaborn:折线图

内容介绍：

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
#加载数据
```

```python
# 网络加载官方案例数据
# tips = sns.load_dataset('tips')

#使用本地数据加载
df = pd.read_csv('seaborn-data/fmri.csv')
df.head()
```

### 1.绘图的一般方法

```python
#阴影部分为置信区间
sns.relplot(x='timepoint', y='signal', data=df,kind='line')
```

```python
#去掉置信区间显示数据
sns.relplot(x='timepoint', y='signal', data=df,kind='line',ci=None)
```

```python
# 增加数据列，并添加不同颜色
sns.relplot(x='timepoint', y='signal', data=df, hue='region', kind='line')
```

```python
# 增加数据列，并添加不同颜色。颜色信息与默认相反。hue_order=['frontal','parietal']参数设置。
sns.relplot(x='timepoint', y='signal', data=df, hue='region', kind='line',hue_order=['frontal','parietal'])
```

```python
#也可添加分图形，按照列数据的个数
sns.relplot(x='timepoint', y='signal', data=df, hue='region', col='event', kind='line')
```

```python
#改变线条央视
sns.relplot(x='timepoint', y='signal', data=df, hue='region', col='event', kind='line',style='region')
```

### 2.使用lineplot绘制折线图

lineplot函数同relplot函数的参数基本一致，但存在下面的差异：

* 不再使用kind参数
* col参数不能使用，即只可以绘制一个图形

```python
sns.lineplot(x='timepoint', y='signal', data=df, hue='region', style='region')
```

### 3.使用类似matplot方法绘制折线图

```python
#绘制简单的折线图
sns.relplot(x='x',
             y='y',
             data=pd.DataFrame({'x':np.arange(10),'y':np.sin(np.arange(10))}),
             kind='line'
            )
```
