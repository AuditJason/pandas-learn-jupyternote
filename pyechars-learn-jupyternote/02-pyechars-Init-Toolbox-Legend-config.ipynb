## pyechars:绘图配置

内容介绍：

```python
from pyecharts.charts import Bar,Scatter
from pyecharts.faker import Faker
from pyecharts import options as opts
```

### 1.InitOpts：初始化配置项

```python
# Faker类中包含一些测试数据
#链式调用
c = (
    Bar(init_opts=opts.InitOpts( width= '400px', height= '300px'))
        .add_xaxis(Faker.choose())
        .add_yaxis('商品', Faker.values())           
    )
c.render_notebook()
```

```python
#正常调用
the_opts = opts.InitOpts( 
    width= '500px', 
    height= '400px',
    bg_color = '#ccc'#背景颜色
)
bar = Bar(init_opts=the_opts )
bar.add_xaxis(Faker.choose())
bar.add_yaxis('商品', Faker.values())
bar.render_notebook()
```

### 2.ToolboxOpts：工具箱配置项

* 标题
* 图例
* 鼠标提示框
* 图像
* 座标点

```python
#正常调用

#设置总体图形参数
all_opts = opts.InitOpts( 
    width= '500px', 
    height= '400px',
)

#初始化图形类
bar = Bar(init_opts=all_opts )
#添加x轴数据
bar.add_xaxis(Faker.choose())
#添加y轴数据
bar.add_yaxis('商品', Faker.values())

#设置标题配置参数
title_opts = opts.TitleOpts(
    title='商品售出情况',
    subtitle=' '*10+'——第二季度情况',  #副标题
    pos_right='20px',    #左边距
    title_textstyle_opts=opts.TextStyleOpts(font_size=12)  #标题字体大小
)

#传入标题参数
bar.set_global_opts(title_opts=title_opts )
bar.render_notebook()
```

### 3.设置图例

```python
#设置总体图形参数
all_opts = opts.InitOpts( 
    width= '500px', 
    height= '400px',
)

#初始化图形类
bar = Bar(init_opts=all_opts )
#添加x轴数据
bar.add_xaxis(Faker.clothes)
#添加y轴数据
bar.add_yaxis('第一季度', Faker.values(),gap='0%') #gap='0%'两个柱状图之间间隙为0
bar.add_yaxis('第二季度', Faker.values(),gap='0%')

#设置图例属性
the_legend_opts = opts.LegendOpts(
    selected_mode='',  #选择模式，可以点击图例后显示/关闭对应的数据。'single'单独显示
    orient='vertical',  #图例是横向还是竖向显示。默认横向。
    is_show = True  #是否展示图例
)

#传入图例参数
bar.set_global_opts(legend_opts=the_legend_opts)
bar.render_notebook()
```
