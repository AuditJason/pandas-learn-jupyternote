## pyechars:条形图绘制

内容介绍：

```python
#导入图形类
from pyecharts.charts import Bar,Scatter
#导入示例数据
from pyecharts.faker import Faker
#导入配置文件
from pyecharts import options as opts
#导入javascript函数接口
from pyecharts.commons.utils import JsCode
```

### 1.x-y轴变换

```python
#正常调用
the_opts = opts.InitOpts( 
    width= '500px', 
    height= '400px',
)
bar = Bar(init_opts=the_opts )
bar.add_xaxis(Faker.choose())
bar.add_yaxis('销售A',Faker.values())
bar.add_yaxis('销售B',Faker.values())
bar.add_yaxis('销售C',Faker.values())
#x-y轴横纵转换
bar.reversal_axis()
#设置图形中数据显示的位置
bar.set_series_opts(label_opts=opts.LabelOpts(position='right'))

#设置图例属性
the_legend_opts = opts.LegendOpts(
    selected_mode='',  #选择模式，可以点击图例后显示/关闭对应的数据。'single'单独显示
    #orient='vertical', #图例是横向还是竖向显示。默认横向。
    pos_top = '5%',#'middle',
    pos_right ='5%',
)

#设置图像标题
bar.set_global_opts(
    opts.TitleOpts(title='条形图-翻转x-y轴'),
    legend_opts=the_legend_opts
)
bar.render_notebook()
```

### 2.堆叠条形图

```python
#正常调用
the_opts = opts.InitOpts( 
    width= '500px', 
    height= '400px',
)
bar = Bar(init_opts=the_opts )
bar.add_xaxis(Faker.choose())
#在数据项中使用stack参数，绘制堆叠条形图
bar.add_yaxis('销售A',Faker.values(), stack='stack1')  #stack=True
bar.add_yaxis('销售B',Faker.values(),stack='stack1')
bar.add_yaxis('销售C',Faker.values(),stack='stack1')
#x-y轴横纵转换
bar.reversal_axis()
#设置图形中数据显示的位置
bar.set_series_opts(label_opts=opts.LabelOpts(position='right'))

#设置图像标题
bar.set_global_opts(
    opts.TitleOpts(title='条形图-堆叠数据'),
    #legend_opts=the_legend_opts
)
bar.render_notebook()
```

### 3.条形图-调节条形间距

```python
#正常调用
the_opts = opts.InitOpts( 
    width= '500px', 
    height= '400px',
)
bar = Bar(init_opts=the_opts )
bar.add_xaxis(Faker.choose())
bar.add_yaxis('销售A',Faker.values(),category_gap='80%')

#设置图形中数据显示的位置
bar.set_series_opts(label_opts=opts.LabelOpts(position='right'))

#设置图像标题
bar.set_global_opts(
    opts.TitleOpts(title='条形图-调节条形间距'),
    legend_opts=the_legend_opts,
)
bar.render_notebook()
```

### 4.背景网状结构

```python
#正常调用
the_opts = opts.InitOpts( 
    width= '500px', 
    height= '400px',
)
bar = Bar(init_opts=the_opts )
bar.add_xaxis(Faker.choose())
bar.add_yaxis('销售A',Faker.values())
bar.add_yaxis('销售B',Faker.values())
bar.add_yaxis('销售C',Faker.values())

#设置图形中数据显示的位置
bar.set_series_opts(label_opts=opts.LabelOpts(position='top'))

#设置图像标题
bar.set_global_opts(
    opts.TitleOpts(title='条形图-背景网状'),
    legend_opts=the_legend_opts,
    xaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=True)),
    yaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=True)),
)
bar.render_notebook()
```
