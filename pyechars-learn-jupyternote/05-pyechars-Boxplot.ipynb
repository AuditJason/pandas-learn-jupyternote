### 2.结合pandas绘图

```python
#使用本地数据加载
df0 = pd.read_csv('athlete_events.csv')
countries = {
    'CHN':'中国',
    'JPN':'日本',
    'KOR':'韩国',
    'USA':'美国',
    'CAN':'加拿大',
    'BRA':'巴西',
    'GBR':'英国',
    'FRA':'法国',
    'ITA':'意大利',
    'ETH':'埃塞俄比亚',
    'KEN':'肯尼亚',
    'NIG':'尼日利亚',
}
df0.head()
```

```python
my_athletes= df0[df0['NOC'].isin(list(countries))]
```

```python
my_athletes.shape
```

```python
my_athletes.dropna(subset=['Height'])#axis='columns',how='all',
```

```python
man_athletes=[]
woman_athletes=[]
for country in countries.keys(): 
    man_athletes.append(my_athletes.query('NOC==@country & Sex=="M"')['Height'])
    woman_athletes.append(my_athletes.query('NOC==@country & Sex == "F"')['Height'])
```

```python
#正常调用
#Boxplot不能使用链式调用
the_opts = opts.InitOpts( 
    width= '500px', 
    height= '400px',
    animation_opts=opts.AnimationOpts(animation=False)  #关掉动画
)
box= Boxplot()#init_opts=the_opts
box.add_xaxis(list(countries.values())),#
box.add_yaxis('男运动员',box.prepare_data(man_athletes))
box.add_yaxis('女运动员',box.prepare_data(woman_athletes))
#设置图像标题
box.set_global_opts(
    title_opts=opts.TitleOpts(title='箱线图-基本示例'),
    yaxis_opts=opts.AxisOpts(interval=20,min_=120,max_=240)  #设置y轴坐标的数据
)
#box.render()
box.render_notebook()
```

### 3.pyechars/seaborn(matplotlib)绘制箱线图比较

1. pyechars绘制的图像能根据图表滑动显示对应数据，方便网页展示。
2. 但 pyechars绘制程序复杂，需要前期的数据处理，且不能直接使用DataFrame数据。 pyechars对部分序列数据格式不支持，如numpy.int。
3. seaborn可以直接使用DataFrame，绘图程序简单，且图像风格更具美感，信息更多。
