## pyechars:VisualMapOpts/TooltipOpts/AxisOpts配置

内容介绍：详见：https://pyecharts.org/#/

```python
from pyecharts.charts import Bar,Scatter
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
```

### 1.VisualMapOpts基本使用

```python
y_values = [(x,y) for x,y in zip(Faker.values(),Faker.values())]
print(y_values)
```

```python
#正常调用
the_opts = opts.InitOpts( 
    width= '500px', 
    height= '400px',
)
bar = Scatter(init_opts=the_opts )
bar.add_xaxis(Faker.clothes)
bar.add_yaxis('销售',y_values)

#设置标题配置参数
viso_opts = opts.VisualMapOpts(
    #type_='size'  # 图形大小。旁边的滑动条表示大于或者小于某值时不展示i
    type_='color',  #根据颜色区分是否展示
    range_text=['大','小']  #滑动条上下的文字
)

#传入标题参数
bar.set_global_opts(visualmap_opts=viso_opts )
bar.render_notebook()
```

### 3.设置轴信息

```python
#正常调用
the_opts = opts.InitOpts( 
    width= '500px', 
    height= '400px',
)
bar3= Bar(init_opts=the_opts )
x_value =[
    '名字很长的X轴标签1',
    '名字很长的X轴标签2',
    '名字很长的X轴标签3',
    '名字很长的X轴标签4',
    '名字很长的X轴标签5',
    '名字很长的X轴标签6',
]
bar3.add_xaxis(x_value )
bar3.add_yaxis('商家A',[10*x for x in range(1,7)])
bar3.add_yaxis('商家B',[10*x for x in range(7,1,-1)])
bar3.render_notebook()
```

```python
#设置轴坐标要素配置信息
the_axis_opts =opts.AxisOpts(
    name='商家名称',
    axislabel_opts = opts.LabelOpts(rotate=-15),
    axisline_opts = opts.AxisLineOpts(symbol='arrow', linestyle_opts=opts.LineStyleOpts(width=2)),  #x轴的箭头
    axistick_opts = opts.AxisTickOpts(is_inside=True,length=20),  #x轴的刻度部分
    axispointer_opts = opts.AxisPointerOpts(is_show=True,type_='line') #鼠标滑动上会有线条
)
bar3.set_global_opts(
    xaxis_opts = the_axis_opts
 )
bar3.render_notebook()
```
