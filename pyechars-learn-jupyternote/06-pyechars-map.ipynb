```python
#导入图形类
from pyecharts.charts import Map
#导入示例数据
from pyecharts.faker import Faker
#导入配置文件
from pyecharts import options as opts
#导入javascript函数接口
from pyecharts.commons.utils import JsCode
```

### 1.中国地图

```python
from pyecharts import options as opts
from pyecharts.charts import Map
from pyecharts.faker import Faker

c = Map()
c.add("淘宝商家商品售卖情况", [list(z) for z in zip(Faker.provinces, Faker.values())], "china")
c.set_global_opts(
    title_opts=opts.TitleOpts(title="Map-基本示例"),
    visualmap_opts=opts.VisualMapOpts()
)
#c.render("map_base.html")
c.render_notebook()
```

### 2.省份地图

```python
c =  Map()
c.add("商家A", [list(z) for z in zip(Faker.guangdong_city, Faker.values())], "广东")
c.set_global_opts(
        title_opts=opts.TitleOpts(title="Map-广东地图"), visualmap_opts=opts.VisualMapOpts()
    )
c.render_notebook()
```

### 3.世界地图

```python
c3 = Map()
c3.add("商家A", [list(z) for z in zip(Faker.country, Faker.values())], "world")
c3.set_series_opts(label_opts=opts.LabelOpts(is_show=False))  #是否显示国名
c3.set_global_opts(
        title_opts=opts.TitleOpts(title="Map-世界地图"),
        visualmap_opts=opts.VisualMapOpts(max_=200),
    )
c3.render_notebook()
```
